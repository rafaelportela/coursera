```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby.gl')
```

```python
products.head()
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
def awesome_count(dict):
    if 'awesome' in dict:
        return dict['awesome']
    else:
        return 0
```

```python
products['awesome'] = products['word_count'].apply(awesome_count)
```

```python
def and_count(dict):
    if 'and' in dict:
        return dict['and']
    else:
        return 0
```

```python
products['and'] = products['word_count'].apply(and_count)
```

```python
def generate_counter(word):
    def words_counter(dict):
        if word in dict:
            return dict[word]
        else:
            return 0
    return words_counter
```

```python
products['and2'] = products['word_count'].apply(generate_counter('and'))
```

```python
products['awesome2'] = products['word_count'].apply(generate_counter('awesome'))
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
selected_words
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(generate_counter(word))
```

```python
products.head()
```

```python
word_usage = {}
for word in selected_words:
    word_usage[word] = products[word].sum()
```

```python
word_usage
```

```python
import operator
word_usage_sorted = sorted(word_usage.items(), key=operator.itemgetter(1))
```

```python
word_usage_sorted
```

```python
products = products[products['rating'] != 3]
```

```python
products['sentiment'] = products['rating'] >= 4
```

```python
products.head()
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
features=selected_words
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                          target='sentiment',
                                                          features=selected_words,
                                                          validation_set=test_data)
```

```python
selected_words_model['coefficients']
```

```python
selected_words_model['coefficients'].sort('value')
```

```python
selected_words_model.evaluate(test_data, metric='roc_curve')
```

```python
graphlab.canvas.set_target('ipynb')
selected_words_model.show(view='Evaluation')
```
